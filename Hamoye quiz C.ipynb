```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import pandas as pd
df=pd.read_csv('Stage C data.csv')
```

```python
df
```

```python
df.drop('stab',axis=1,inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X=df.drop('stabf',axis=1)
y=df['stabf']
```

```python
X,y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
scaled_train=pd.DataFrame(sc.fit_transform(X_train),columns=X_train.columns)
scaled_test=pd.DataFrame(sc.transform(X_test),columns=X_test.columns)
```

```python
scaled_train
```

```python
scaled_test
```

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
rf=RandomForestClassifier(random_state=1)
ext= ExtraTreesClassifier(random_state=1)
xgb=XGBClassifier(random_state=1)
lgb=LGBMClassifier(random_state=1)
```

```python
rf.fit(scaled_train,y_train)
ext.fit(scaled_train,y_train)
xgb.fit(scaled_train,y_train)
lgb.fit(scaled_train,y_train)
```

```python
pred_rf=rf.predict(scaled_test)
pred_ext= ext.predict(scaled_test)
pred_xgb= xgb.predict(scaled_test)
pred_lgb= lgb.predict(scaled_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
np.round(accuracy_score(y_test,pred_rf),4)
print('Random Forest Accuracy: {}' .format(accuracy_score(y_test,pred_rf)))
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
np.round(accuracy_score(y_test,pred_xgb),4)
print('Xgboost Accuracy: {}' .format(accuracy_score(y_test,pred_xgb)))
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,confusion_matrix
np.round(accuracy_score(y_test,pred_lgb),4)
print('lightgbm Accuracy: {}' .format(accuracy_score(y_test,pred_lgb)))
```

```python
#Hyperparameter grig for Randomizedsearchcv on ExtraTreeClassifier
n_estimators=[50,100,300,500,1000]
min_samples_split=[2,3,5,7,9]
min_samples_leaf=[1,2,4,6,8]
max_features=['auto','sqrt','log2', None]


hyperparameter_grid= {'n_estimators':n_estimators,'min_samples_leaf':min_samples_leaf,
                      'min_samples_split':min_samples_split,'max_features':max_features}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rcv= RandomizedSearchCV(ext, hyperparameter_grid,cv=5,random_state=1,scoring='accuracy',n_iter=10,n_jobs=-1,verbose=1)
rcv.fit(scaled_train,y_train)
rcv.best_params_
```
